```python
import pandas as pd
import faiss
import pickle
from sentence_transformers import SentenceTransformer
```

```python
# Load insights CSV
df = pd.read_csv("C:/Users/ritik/Downloads/telecom_churn_insights.csv")
```

```python
print(df)
```

```python
# Convert insights into list
texts = df["Insight"].tolist()
```

```python
# Embedding model
model = SentenceTransformer(
    "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# Create embeddings
embeddings = model.encode(texts)
```

```python
# Create FAISS index
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)
```

```python
# Save
faiss.write_index(index, "telecom_index.faiss")

with open("telecom_texts.pkl", "wb") as f:
    pickle.dump(texts, f)

print("Vector DB Created")
```
